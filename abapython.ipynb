```python
import csv

# Function to create a CSV file with headers
def create_csv(filename, headers):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)

# Function to append rows to a CSV file
def append_row(filename, row):
    with open(filename, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(row)

# Example usage
filename = 'python.csv'
headers = ['Name_laptop', 'Price', 'Ram','OS','Processor','Rating', 'Number of Rating','Reviews','Display_inch','Storage']

# Create the CSV file with headers
create_csv(filename, headers)

# Append rows to the CSV file
data = [
    ['HP 15s Intel Core i5 12th Gen 1235U - (16 GB/512 GB SSD/Windows 11 Home) 15s-fq5112TU Thin and Light Laptop(15.6 inch, Jet Black, 1.69 Kg, With MS Office', 56499,16,'Windows11Home','Core i5',4.1,293,27,15.6,514],
    ['HP Omen AI Powered AMD Ryzen 7 Octa Core 7840HS - (16 GB/1 TB SSD/Windows 11 Home/6 GB Graphics/NVIDIA GeForce RTX 4050) XD0007ax Gaming Laptop(16.1 Inch, Shadow Black, 2.32 Kg)',101990,16,'Windows11Home','Ryzen 7 Octa Core',4.6,16,1,16.1,1],
['HP Laptop Intel Core i3 11th Gen 1115G4 - (8 GB/512 GB SSD/Windows 11 Home) 15s-fq2717TU Thin and Light Laptop(15.6 Inch, Natural Silver, 1.69 Kg, With MS Office)',34990,8,'Windows11Home','Core i3',4.2,1006,91,15.6,512],
['New Samsung Galaxy Book Go 5G 14” 128GB 8GB RAM Wifi + Cellular Unlocked Laptop',20358,8,'Windows 11','Qualcomm Snapdragon 8cx Gen 2',3.8,1059,24,14,128]
]

for row in data:
    append_row(filename, row)

print("Rows appended successfully.")
```

```python
num_attr, cat_attr = [], []

for col in df.columns:
    if df[col].dtype == "object":
        cat_attr.append(col)
    else:
        num_attr.append(col)

print(num_attr, cat_attr, sep = "\n")
```

```python
for col in cat_attr:
    print(f"Number of unique values in {col}: {df[col].nunique()}")
    print(df[col].value_counts(), "\n\n")
    if (df[col].nunique() <= 10):
        plt.pie(df[col].value_counts(), autopct = "%.2f%%", labels = np.unique(df[col]))
        plt.show()
    else:
        plt.figure(figsize = (18, 3))
        sns.countplot(x = df[col])
        plt.xticks(rotation = 90)
        plt.show()
```

```python
by_brand = df.groupby(["Name_laptop"])["Price"].mean().sort_values(ascending = False)
print(by_brand)
```
